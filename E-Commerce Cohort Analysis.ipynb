```python
import pandas as pd
import numpy as np
import datetime as dt
```

```python
#input data
df = pd.read_csv('Downloads/e-commerce/data.csv', engine='python', encoding= 'ISO-8859-1')
df
```

```python
df.info()
```

```python
import dateutil
from datetime import datetime as dt
from pytz import utc
```

```python
df['datetime'] = df['InvoiceDate'].apply(lambda x: dateutil.parser.parse(x).timestamp())
df['month'] = df['datetime'].apply(lambda x: dt.fromtimestamp(x, utc).month)
df['year'] = df['datetime'].apply(lambda x: dt.fromtimestamp(x, utc).year)
```

```python
df.head()
```

```python
df['cohort'] = df.apply(lambda row: (row['year'] * 100) + (row['month']), axis=1)

cohorts = df.groupby('CustomerID')['cohort'].min().reset_index()
cohorts.columns = ['CustomerID', 'first_cohort']

df = df.merge(cohorts, on='CustomerID', how='left')
```

```python
cohorts.head()
```

```python
headers = df['cohort'].value_counts().reset_index()
headers.columns = ['Cohorts', 'Count']
headers.head()
headers = headers.sort_values(['Cohorts'])['Cohorts'].to_list()
```

```python
headers
```

```python
df.dropna(inplace=True)
```

```python
df['cohort_distance'] = df.apply(lambda row: (headers.index(row['cohort']) - headers.index(row['first_cohort'])) if (row['first_cohort'] != 0 and row['cohort'] != 0) else np.nan, axis=1)
```

```python
df.head()
```

```python
cohort_pivot = pd.pivot_table(df,
                              index='first_cohort',
                              columns='cohort_distance',
                              values='CustomerID',
                              aggfunc=pd.Series.nunique)
```

```python
cohort_pivot
```

```python
cohort_pivot_pers = cohort_pivot.div(cohort_pivot[0],axis=0)
```

```python
cohort_pivot_pers
```

```python
cohort_size = cohort_pivot.iloc[:, 0]
cohort_size
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

fig_dims = (12, 8)
fig, ax = plt.subplots(figsize=fig_dims)

sns.heatmap(cohort_pivot_pers, annot=True, fmt='.0%', mask=cohort_pivot.isnull(), ax=ax, square=True, linewidths=.5, cmap=sns.cubehelix_palette(start=2))

plt.show()
```
